```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/zillow_data.csv') #upload data
df

#df = df.drop("State", axis=1)
#df = df.drop("RegionType", axis=1)


```

```python
copy_df = df.copy()
copy_df.shape
copy_df.dtypes
```

```python
copy_df['RegionName'].unique() #finding unique variables in the array
```

```python
duplicates = copy_df.duplicated().sum() #finding the total duplicates in the dataset
duplicates
copy_df = copy_df.drop(columns= ['SizeRank', 'RegionID','StateName', 'RegionType', 'Metro', 'CountyName'] ) #drop unnecessary columns
copy_df
```

```python
copy_df["City"] = copy_df['RegionName'] + ' , ' + copy_df['State']
copy_df
cols = ['City'] + [col for col in copy_df.columns if col != 'City']
copy_df = copy_df[cols]
copy_df
```

```python
t_data = copy_df.T #transposing data
t_data.to_csv("t_data.csv", index =False)
t_data = t_data.drop(index = t_data.index[1:3])
t_data
```

```python
t_data.dtypes
```

```python
t_data.isnull().sum()
```

```python
t_data.columns = t_data.iloc[0] #set first row as column header
#t_data
t_data = t_data.drop(index= t_data.index[0:1])
t_data
```

```python
t_data = t_data.reset_index()
t_data
```

```python
t_data.rename(columns = {'index' : 'Date'}, inplace=True)
t_data
```

```python
t_city = t_data[['Date', 'New York , NY']]
t_city
```

```python
t_city['New York , NY'] = t_city['New York , NY'].astype(float)
```

```python
t_city.dtypes
```

```python
#Convert to datetime
t_city['Date'] = pd.to_datetime(t_city['Date'])
t_city
```

```python
t_city.fillna(t_city.mean(numeric_only= True))
t_city
```

```python
t_city.dtypes
```

```python

#Create the line plot
plt.figure(figsize=(12, 6)) #creates placeholder for any chart or line chart
plt.plot(t_city['Date'], t_city['New York, NY'], label = 'New York Housing Prices') #you want to plot between date and newyork column and want to label it ny housing prices
plt.xlabel('Date')
plt.ylabel('Housing Prices')
plt.title('New York Housing Prices Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
t_city['Year'] = t_city['Date'].dt.year
t_city['Month'] = t_city['Date'].dt.month
t_city['Month_Year'] = t_city['Year'] * 100 + t_city['Month']

print(t_city.head())
```

```python
#Define features and target
#Define Features (X)
X = t_city[['Month_Year']]
y = t_city[['New York , NY']]

#Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

#fit to the Linear Regression Model
pricepredictor = LinearRegression()
pricepredictor.fit(X_train, y_train)
```
